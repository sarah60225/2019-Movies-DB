# Cast Table---
## Extract  casts (including actors, directors,..) corresponding to movies


```python


cast=[]
job=[]
movie_2=[]

t1=' (director)'
t2=' (screenplay)'
t3=' (director/screenplay)'
t4=' (co-director)'
t5=' (directors)'
t6=' (screenwriter)'
# Split cast into different rows
for i, act in enumerate(movies['Cast and crew']):
    cast_list = re.split(";\s|,\s",act)  ## split cast list for each movie into several rows
    
    for j, d in enumerate(cast_list):
        if t1 in d:
            d= d.rstrip(t1)  # remove blank on the left of dirctor name
            job.append("director")
        elif t2 in d :
            d= d.rstrip(t2)
            job.append("screenplay")
        elif t3 in d:
            d= d.rstrip(t3)
            job.append("director/screenplay")
        elif t4 in d:
            d= d.rstrip(t4)
            job.append("co-director")
        elif t5 in d:
            d= d.rstrip(t5)
            job.append("director")
        elif t6 in d:
            d= d.rstrip(t6)
            job.append("screenplay")
        else:
            job.append("actor")
         
       
        movie_2.append(movies['Title'][i])
        cast.append(d)
```

```python
df_2 = pd.DataFrame(data = cast, columns = ['Name'])

df_2['Job'] = job
df_2['Movie'] = movie_2
```

```python
## Again, split director/screenplay into diffrent rows

cast_id = []
cast_21=[]
job_21=[]
movie_21=[]

for i, job in enumerate(df_2['Job']):

    job_list = job.split("/")  ## split director/screenplay into diffrent rows
    
    for j, jjob in enumerate(job_list):
        
        cast_21.append(df_2['Name'][i])
        job_21.append(jjob)
        movie_21.append(df_2['Movie'][i])
```

```python
df_21 = pd.DataFrame(data = cast_21, columns = ['Name'])

df_21['Job'] = job_21
df_21['Movie'] = movie_21
```

```python
## Add cast_id
for i, movie in enumerate(movies['Title']):

    ## Creat starting no. of cast_id corresponding to movie
                                
        array = movies[movies['Title'] == movie].index.values
        start_num=array[0]
        row = df_21[df_21['Movie'] == movie].index.values
    
        for j in range(0,len(row)):    ## Creat endding no. of cast_id from the count of each movie from df_21                    
            aa=str(start_num+1)
            bb=str(j+1)
            c=aa.zfill(3)
            d=bb.zfill(2)
        
            cast_id.append(str(c+d))
df_21['cast_id'] = cast_id
df_21 = df_21.reindex(columns=['cast_id', 'Name','Job', 'Movie']) 

df_21.head(5)
```

```python
## Export the data to cast_table
df_21.to_csv('cast_table.csv', index=False)
```
